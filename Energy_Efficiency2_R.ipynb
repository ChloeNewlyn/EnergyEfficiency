**How to import the dataset in Google Colab? (3 Mark) How to use mathematical expressions/ aggregate functions/ arithmetic functions to query and manipulate data?**

We can use mathematical expressions, aggregate functions and arithmetic functions to help us to gain new insights into our data and manipulate the information for a specific outcome to gain new knowledge.

Mathematical Expressions: In sql, you can multiply (*), subtract (-), add (+),divide(/) data and much more. For instance, in the below statement, if we did not have the surface area of our houses, we could calculate this using the wall area and the roof area to gain the surface area. This uses the addition mathematical expression sql statement.

```r
result <- sqldf("SELECT Wall_Area+ Roof_Area AS Surface_Area From energy_data ORDER BY Surface_Area ASC")
print(result)
```

Aggregate functions: In sql, you can find the average (AVG), sum (SUM), maximum (MAX), minimum (MIN), and count (COUNT) of a set of data. This helps us to understand any trends, insights or anomalies.
The below simple Query uses the AVG aggregate function to demonstrate how this can be used. The results show the average heating and cooling load for the houses. The result is that they are very similar, but the cooling load is higher compared to the heating load.

```r
result <- sqldf("SELECT AVG(Heating_Load), AVG(Cooling_Load) FROM energy_data")
print(result)
```

If we wish to see how many houses have a lower heating and cooling load compared to average, we can use the below query. We can see that 393 houses have a heating and cooling load less than the average.

```r
result <- sqldf("SELECT COUNT (*) FROM energy_data WHERE Heating_Load<22 AND Cooling_Load<24")
print(result)
```

To look at the min and max of both heating and cooling load, we can do the following:

```r
result <- sqldf("SELECT MAX(Heating_Load), MIN(Heating_Load), MAX(Cooling_Load), MIN (Cooling_Load)FROM energy_data")
print(result)
```

We can use this data to calculate the range of data and we can see that the min and max for both have only a minimal difference in range.

```r
result <- sqldf("SELECT MAX(Heating_Load) - MIN(Heating_Load), MAX(Cooling_Load) -MIN (Cooling_Load)FROM energy_data")
print(result)
```

You can use the SUM function to see the sum of the whole column.

```r
result <- sqldf("SELECT SUM(Heating_Load) AS total_heating_load, SUM(Cooling_Load) AS total_cooling_load FROM energy_data")
print(result)
```

**Executing SQL Queries in R: (4 Marks) How do you execute SQL queries in R?
Can you provide three examples of SQL queries executed within R?**

With this data, we can use mathematical expressions to understand and manipulate the data to gain valuable information . The first query is seeing the volume of the houses. As we have the area of the wall, roof and surface, we can see the whole volume of the area by multiplying these areas together. This can allow us to see if the volume of the area could potentially be affecting the coolling and heating load. If we put this in ascending order by the total volume, we can see that places with the same volume do not all have the same loads, so the total volume is probably not affecting the load.




```r
result <- sqldf("SELECT Surface_Area * Wall_Area * Roof_Area AS Total_Volume, Heating_Load, Cooling_Load From energy_data ORDER BY Total_Volume ASC")
print(result)
```

With this information, we can do simple queries. If you wish to narrow down a search, you can use the OR statement to bring up results that can be applied to one or the other criteria. Here the criteria is to have a certain volume (37059435) or have a heating load of 29.9 and be in ascending order of heating to see where the data falls and if there is any patterns.

```r
result <- sqldf("SELECT Surface_Area * Wall_Area * Roof_Area AS Total_Volume, Heating_Load, Cooling_Load From energy_data WHERE Total_Volume=37059435 OR Heating_Load=29.9 ORDER BY Heating_Load ASC")
print(result)
```

We can also use GROUP BY function to help us to understand data between groups. In the following, I have seen the average,max and min heating load of the groups of relative compactness. We can see that the higher relative compactness in a group does not mean a higher average load and that the group 0.79 has the highest average.

```r
result <- sqldf("SELECT Relative_Compactness, MAX(Heating_Load) AS Max_Load, AVG(Heating_Load) AS Avg_Load, MIN(Heating_Load) AS Min_Load FROM energy_data Group By Relative_Compactness")
print(result)
```

**Data Visualization in R: (3 Marks)
How do you create informative and visually appealing plots using the R package for data visualization? Provide three examples of different types of plots that can be created using R and explain their significance in data analysis.**

To create visual plots using R, we must install ggplot2 and then call that package.  

The first plot I will create is a Histogram.




```r
install.packages("ggplot2")
library(ggplot2)
```

To create a Histogram with the heating load, we must do the following to create a histogram with our energy_data.This will visually show us the frequency of the same heating loads.

*Histogram significance:* Using a histogram will allow us to visually see the spread and frequency of the data, see if there are any trends, if the data is skewed one way or the other or if there are any anomolies. For instance, we can see from our histogram, that one heating load bin at around 14/15 has a high frequency of 50+ which means that this heating load amount is shown a lot more in the data compared to others. You can then use this to investigate as to why that is or if there is any similar factors that will help to predict whether the houses in the future will also have the same heating load.

```r
library(ggplot2)
ggplot(energy_data, aes(x=Heating_Load)) +
    geom_histogram(binwidth= 0.5, fill="Pink", color="black") +
      labs(title="Histogram of Heating Load", x="Heating Load", y="Frequency")
```

The next plot created is a scatter plot.

*Scatter Plot Significance:* A scatter plot allows you to see the relationship/correlation between two variables and see any trends to help with predictions. For instance, in the plot below we can see that a bigger surface area usually indicates a smaller cooling load, but once it gets to around 650, the cooling load massively increases and then around the 600 and lower surface area has a cooling load that lays at the mid range. This can help to use the surface area to predict future cooling loads using this trend.

```r
scatter_plot <- ggplot(energy_data, aes(x = Cooling_Load , y = Surface_Area)) +
  geom_point()+
  labs(title = "Relationship between Cooling Load and Surface Area",
       x = "Cooling_Load",
       y = "Surface_Area") +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
print(scatter_plot)
```

The last plot is a bar chart. In the below chart, I needed to aggregate my data by using the group by function for the relative compactness and then using the Mean function and calculating the standard deviation and using summarise so that these can be calculated.I called the dyplr package so that I could piece the operations. The output shows the heating load by relative compactness.

*Bar Chart Significance:* Using a bar chart allows to see which groups have the highest and lowest amount against other groups to see trends. For example, we can see that relative compactness group near 0.8. has the highest mean heating_load in comparison to the other groups. This means that we could possibly make predicitions about future data and any relative compactness that falls in this group, will probably have a mean heating load also at this level.

```r
library(dplyr)
Aggregate_data <- energy_data %>%
  group_by(Relative_Compactness)%>%
  summarise(Mean_Heating_Load = mean(Heating_Load),
sd_Heating_Load = sd(Heating_Load))

platform_bar <- ggplot(Aggregate_data, aes(x = Relative_Compactness, y = Mean_Heating_Load)) +
  geom_bar(stat = "identity", fill = "Pink") +
  labs(title = "Heating Load by Relative Compactness", x = "Relative Compactness", y = "Mean Heating Load") +
  theme_minimal()
print(platform_bar)
```
